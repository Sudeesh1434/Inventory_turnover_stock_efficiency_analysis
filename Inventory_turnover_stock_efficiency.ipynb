```python
#Create Average Inventory
df['average_inventory']=(df['opening_stock']+df['closing_stock']/2)
df['average_inventory']
```

```python
#Product-wise Inventory Turnover
product_turnover=df.groupby('product_name').agg(
    total_sold=('units_sold','sum'),
    avg_turnover=('average_inventory','mean'))
product_turnover['inventory_turnover']=(
    product_turnover['total_sold']/product_turnover['avg_turnover'])
product_turnover
```

```python
#Identify Slow-Moving Products
slow_moving=product_turnover[
product_turnover['inventory_turnover']<1
]
slow_moving
```

```python
#Blocked Capital in Inventory
df['blocked_value']=df['closing_stock']*df['unit_cost']
blocked_capital=(df.groupby('product_name')['blocked_value'].sum()
                .sort_values(ascending=False))
blocked_capital
```
